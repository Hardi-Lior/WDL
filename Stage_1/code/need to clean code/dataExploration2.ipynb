```python
data = df2
#graph = sns.FacetGrid(data, col="County_of_Public_Transportation")
```

```python
#graph.map(sns.histplot, "County_of_Origin")
g = sns.FacetGrid(data, col="GenderDescription", hue="District_of_Origin")#
g.map(sns.barplot, "Average_BusUsers_per_Day", "AgeClassDescription", alpha=1)
g.add_legend()
```

```python
g = sns.FacetGrid(data, row="smoker", col="time", margin_titles=True)
g.map(sns.regplot, "size", "total_bill", color=".3", fit_reg=False, x_jitter=.1)
```

```python
g = sns.PairGrid(data)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False)
```

```python
g = sns.PairGrid(data, vars=["sepal_length", "sepal_width"], hue="species")
g.map(sns.scatterplot)
```
